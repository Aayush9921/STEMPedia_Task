```python
def ExtractCbse(list):    
    for i in list:
        url='https://www.prokerala.com/education/cbse-schools-in-'+i+".html"
        gujrat=requests.get(url)
        soup=BeautifulSoup(gujrat.text,"html.parser")
        school_list = soup.find('ul', class_='grid-col grid-col-xs-12 grid-col-sm-6 f-sans-serif t-small')
        if school_list:
            for school_item in school_list.find_all('li', class_='ellipsis pad-large'):
                school_name = school_item.find('a').text.strip()
                school_link = school_item.find('a')['href']
                address_span = school_item.find('span', class_='text-muted t-small')
                address = address_span.text.strip() if address_span else None 
                phone_span = school_item.find('span', class_='text-muted t-xsmall')
                phone = phone_span.text.strip() if phone_span else None
    
                CBSEData.append({
                        'School Name': school_name,
                        'School Address': address,
                        'School Phone': phone,
                        'affiliation':'CSEB'
                    })
    
```
